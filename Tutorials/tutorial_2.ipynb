## Using word vectors from spaCy

spaCy offers pretrained vectors for each token (you can check <a href="https://spacy.io/usage/vectors-similarity">here</a> for more detials). They come inside the models that we load. However, the small model doesn't include vectors. Let's look at what happens...

```python
import spacy
# To download the model comment in one of the following lines:

#!python3 -m spacy download en_core_web_sm
# or
# spacy.cli.download("en_core_web_sm")
```

```python
from spacy.lang.en import English
nlp = English()

nlp = spacy.load('en_core_web_sm')
```

```python
text = "superman batman cat banana"

tokens = nlp(text)
for token in tokens:
    print(token.text, token.has_vector, token.vector_norm, token.is_oov) # oov = Out of Vocabulary
```

```python
for token1 in tokens:
    for token2 in tokens:
        print(token1.text, token2.text, token1.similarity(token2))
```
